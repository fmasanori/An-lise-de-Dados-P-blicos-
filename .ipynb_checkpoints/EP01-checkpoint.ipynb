```python
import os
import pandas as pd
cities=[]
files=os.listdir('database/csv')
for f in files:
    x=f.split('.')
    if x[1]=='csv':
        df=pd.read_csv('database/csv/'+f)
        cities.append([x[0],df])
```

# Análise dos Profissionais da área da Saúde no Vale do Paraíba

```python
import matplotlib.pyplot as plt
for i in cities:
    name=i[0].replace('_',' ')
    i[1].plot(figsize=(15,5),x='Categoria', y=['Profissionais/1000 habitantes','Profissionais pelo SUS/1000 habitantes'],kind='bar')
    plt.title(name)
    plt.show()
```

```python
data={}
cat=[]
for c in cities[0][1]['Categoria']:
    cat.append(c)
ts=[]
tns=[]
for x in range(19):
    total_sus=0
    total_nsus=0
    for tbl in cities:
        total_sus+=tbl[1]['Atendem pelo SUS'][x]
        total_nsus+=tbl[1]['Não atendem pelo SUS'][x]
    ts.append(total_sus)
    tns.append(total_nsus)
data['Categoria']=[ct for ct in cat]
data['Atendem pelo SUS']=[t for t in ts]
data['Não atendem pelo SUS']=[ y for y in tns]
total=pd.DataFrame.from_dict(data)
total.plot(figsize=(15,6), x='Categoria', y=['Atendem pelo SUS','Não atendem pelo SUS'],kind='bar')
plt.title('Total de profissionais na área da saúde no Vale do Paraíba')
plt.show()
```
